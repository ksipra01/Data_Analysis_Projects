```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("student_result_analysis.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
```

```python
print(df.head())
```

```python
df.head()
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df,x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chat we have analysed that:
#the number of females in the data is more  than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb,annot = True)
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their studies
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb1,annot = True)
plt.show()
```

```python
#from the above chart we analyzed that their is negligible imapct on the student's score due to their marital status
```
